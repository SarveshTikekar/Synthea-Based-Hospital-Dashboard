```python
for index, col in enumerate(data.columns):
    print(f"Index: {index}, col name: {col} \n")
```

```python
data.iloc[:,0] = (pd.to_datetime(data.iloc[:,0]) + pd.Timedelta(hours=5, minutes=30)).dt.strftime("%Y-%m-%dT%H:%M:%S")
data.iloc[:,1] = (pd.to_datetime(data.iloc[:,1]) + pd.Timedelta(hours=5, minutes=30)).dt.strftime("%Y-%m-%dT%H:%M:%S")

data
```

```python
new_cols = ['medication_start', 'medication_end', 'uuid', '_', 'encounter_id','rxnorm_medication_id', 'medication', 'medication_cost', 'amount_covered_under_claim', 'quantity', 'net_cost', 'condition_id', 'condition_treated']
old_cols = data.columns

for new, old in zip(new_cols, old_cols):
    data.rename(columns={old: new}, inplace=True)

data.drop(inplace=True, columns=[col for col in data.columns if col == "_"])
data
```
