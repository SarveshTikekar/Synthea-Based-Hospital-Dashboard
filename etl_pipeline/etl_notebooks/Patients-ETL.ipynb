```python
#KPI-1

patient_count = len(data["uuid"])
patient_count
```

```python
#KPI-2 Active patient rate

active_patient_count = int((data["death_date"].isna().sum() / patient_count) * 100)
active_patient_count
```

```python
#KPI-3 Male to female ratio
gender_data = dict(data["gender"].value_counts())
male_to_female_ratio = int((gender_data.get("M", 0) / gender_data.get("F", 1)) * 100)
print(male_to_female_ratio)
```

```python
#KPI-4 Mean and median family income
mean_fi, median_fi = int(data[data["death_date"].isna()][["family_income"]].mean().iloc[0]), int(data[data["death_date"].isna()][["family_income"]].median().iloc[0])

print(type(mean_fi))
print(mean_fi, median_fi, sep="\n")

# alive_patients_family_income = data[data["death_date"].isna()]
```

```python
#Metric-1

data["birth_date"] = pd.to_datetime(data["birth_date"])
data["death_date"] = pd.to_datetime(data["death_date"])

current_date = pd.Timestamp.now()
data["end_date"] = data["death_date"].fillna(current_date)
data["age"] = ((data["end_date"] - data["birth_date"]).dt.days / 365.25).astype(int)

econ_dep_ratio = 0
working_age_pop = len(data[(data["age"] >= 20) & (data["age"] <= 64)])

# Calculate the ratio
dependent_pop = len(data[data["age"] < 20]) + len(data[data["age"] > 64])

# Prevent division by zero
if working_age_pop > 0:
    econ_dep_ratio = (dependent_pop / working_age_pop) * 100
else:
    pass

print(econ_dep_ratio)
```
